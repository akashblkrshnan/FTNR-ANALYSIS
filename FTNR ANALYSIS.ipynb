# Branch and Agent wise analysis

```python
import pandas as pd  

file_path = "C:\\Users\\Akash\\Desktop\\FTNR-PROJECT\\Nov'24-FTNR_DA_Project\\FTNR Analysis_2024.xlsx"  
data = pd.read_excel(file_path)

def determine_ftnr_status(row):
    # If it's the first attempt and DDE Status is "Approved," it's FTR
    if row['Attempt'] == 1 and row['DDE_Status'] == "Approved":
        return "FTR"
    # If Previous Query is blank, mark as "GOOD FTNR"
    elif pd.isna(row['Previous Query']):
        return "GOOD FTNR"
    # If Previous Query matches the current Query, mark as "GOOD FTNR"
    elif row['Previous Query'] == row['Query']:
        return "GOOD FTNR"
    # Otherwise, it's "BAD FTNR"
    else:
        return "BAD FTNR"

data['FTNR Status (FINAL)'] = data.apply(determine_ftnr_status, axis=1)

agent_group = data.groupby('userid')['FTNR Status (FINAL)'].value_counts(normalize=True).unstack(fill_value=0)
agent_group = agent_group * 100  
agent_group = agent_group.reset_index()  

agent_group.rename(columns={
    "GOOD FTNR": "Good FTNR %",
    "BAD FTNR": "Bad FTNR %",
    "FTR": "FTR %"
}, inplace=True)


branch_group = data.groupby('Branch')['FTNR Status'].value_counts(normalize=True).unstack(fill_value=0)
branch_group = branch_group * 100  
branch_group = branch_group.reset_index()  

branch_group.rename(columns={
    "GOOD FTNR": "Good FTNR %",
    "BAD FTNR": "Bad FTNR %",
    "FTR": "FTR %"
}, inplace=True)


output_file = "C:\\Users\\Akash\\Desktop\\FTNR-PROJECT\\Nov'24-FTNR_DA_Project\\FTNR Analysis_2024.xlsx"  
with pd.ExcelWriter(output_file) as writer:
    
    data.to_excel(writer, sheet_name="FTNR Status", index=False)
    agent_group.to_excel(writer, sheet_name="Agent Analysis", index=False)

    branch_group.to_excel(writer, sheet_name="Branch Analysis", index=False)

print("The FTNR analysis is complete! Results have been saved to a new Excel file.")
```

```python
import pandas as pd

file_path = "C:\\Users\\Akash\\Desktop\\FTNR-PROJECT\\Nov'24-FTNR_DA_Project\\FTNR Analysis_2024.xlsx"
data = pd.read_excel(file_path)

def determine_ftnr_status(row):
    # If it's the first attempt and DDE Status is "Approved," it's FTR
    if row['Attempt'] == 1 and row['DDE_Status'] == "Approved":
        return "FTR"
    # If there is no new query, it's GOOD FTNR
    if pd.isna(row['Previous Query']):
        return "GOOD FTNR"
    # If the current query matches the previous query, it's GOOD FTNR
    if row['Query'] == row['Previous Query']:
        return "GOOD FTNR"
    # Otherwise, it's BAD FTNR
    return "BAD FTNR"

data['FTNR Status (FINAL)'] = data.apply(determine_ftnr_status, axis=1)

# Agent-wise FTNR analysis
agent_group = data.groupby('username')['FTNR Status (FINAL)'].value_counts(normalize=True).unstack(fill_value=0) * 100
agent_group.rename(columns={
    "GOOD FTNR": "Good FTNR %",
    "BAD FTNR": "Bad FTNR %",
    "FTR": "FTR %"
}, inplace=True)

# Branch-wise FTNR analysis
branch_group = data.groupby('Branch')['FTNR Status (FINAL)'].value_counts(normalize=True).unstack(fill_value=0) * 100
branch_group.rename(columns={
    "GOOD FTNR": "Good FTNR %",
    "BAD FTNR": "Bad FTNR %",
    "FTR": "FTR %"
}, inplace=True)

# Display FTNR Status Data
print("FTNR Status Analysis:")
display(data)

# Display Agent-Wise Analysis
print("\nAgent-Wise FTNR Analysis:")
display(agent_group)

# Display Branch-Wise Analysis
print("\nBranch-Wise FTNR Analysis:")
display(branch_group)
```
